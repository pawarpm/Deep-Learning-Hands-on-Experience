## 1 — Utility: set reproducible seeds

```
import random, numpy as np, tensorflow as tf, os
def set_seeds(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
set_seeds(42)
print('Seeds set for reproducibility')
```
