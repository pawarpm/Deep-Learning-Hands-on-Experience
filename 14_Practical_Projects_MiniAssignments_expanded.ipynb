## 1 — Project: Flower Classifier (Transfer Learning)

**Dataset:** Oxford Flowers 102 (link: download via [official site](https://www.robots.ox.ac.uk/~vgg/data/flowers/102/))

**Deliverables:** notebook, saved model, README with dataset & training details, accuracy/confusion matrix.

**Suggested steps:**
1. Prepare dataset in folders `flowers/<class>/*.jpg`.
2. Use `image_dataset_from_directory` to load.
3. Fine-tune MobileNetV2 or ResNet50.
4. Save model and produce sample predictions.

## 2 — Project: Sugarcane Disease Detector (Object Detection)

**Dataset:** collect 150–400 annotated images. Use VIA or Roboflow.

**Deliverables:** YOLOv8 weights, notebook, mAP plot. Use Roboflow export for YOLO format or VIA converted to YOLO.

**Suggested steps:**
- Annotate images in Roboflow or VIA.
- Export YOLOv8 format and train with `ultralytics`.
- Evaluate and produce PR curve and mAP summary.

## 3 — Project: Bird-call Audio Classifier

**Dataset:** ESC-50 subset or BirdCLEF sample.

**Deliverables:** spectrogram-based CNN model, confusion matrix, audio demo of predictions.

**Suggested steps:**
- Convert wav files to mel-spectrograms.
- Train small CNN and evaluate.
- Provide sample audio inputs and predicted labels.

## 4 — Project: NDVI Time Series Forecast

**Dataset:** monthly NDVI CSV for a farm/plot (3+ years recommended).

**Deliverables:** forecast notebook, plots, MAPE metric, saved model.

**Suggested steps:**
- Feature engineering (lags, rolling means).
- Train LSTM/TCN/Transformer and compare baseline.
- Provide forecast horizon examples and error metrics.
